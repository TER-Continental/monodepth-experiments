```python
from monodepth import Monodepth

import matplotlib.pyplot as plt
from PIL import Image
```

```python
frame = Image.open("monodepth2/assets/test_image.jpg").convert('RGB')
```

```python
model = Monodepth("mono_640x192")
im, disparity = model.parse_frame(frame)
```

```python
plt.subplot(1, 2, 1)
plt.imshow(frame)
plt.subplot(1, 2, 2)
plt.imshow(im)

plt.tight_layout()
plt.show()
```
